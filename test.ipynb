```
import pandas as pd
import nltk
from pandas import ExcelWriter
from pandas import ExcelFile
from spellchecker import SpellChecker 
nltk.download('popular')
stop = nltk.corpus.stopwords.words('english')
words = nltk.corpus.words.words()
spell = SpellChecker()
data = pd.read_excel('/content/sample_data/anger-train.xlsx',header=None ,delimiter='\t', dtype=str, names=['tweet'])
print(data.head())
#Still need to be improved
data['tweet'] = data['tweet'].apply(lambda x: x.lower())
data['tweet'] = data['tweet'].apply(lambda x: " ".join([''.join([j for j in i if not j.isdigit() and j.isalpha()]) for i in x.split(' ')[:-4]]))
#should I correct spellings? data['tweet'] = data['tweet'].apply(lambda x: " ".join([spell.correction(i) for i in x.split(' ') ]))
#maybe sentences lose meaning if i only keep right words
data['tweet'] = data['tweet'].apply(lambda x: " ".join(i for i in x.split(' ') if i in words and not i in stop))



pd.set_option('display.max_colwidth',1000)
pd.set_option('display.max_rows',1000)
print(data['tweet'])
data.to_excel('/content/sample_data/clean-anger-train.xlsx')
```
